```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('income2.csv')
df.columns = df.columns.str.strip()
# Plotting
plt.xlabel('Year')
plt.ylabel('Income')

# Scatter plot
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], color='red', marker='+')  # Access columns using iloc

# Show the plot
plt.show()
```

```python

new_df = df.drop('income',axis='columns')
new_df
```

```python
price = df.income
price
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

```python
reg.predict([[2020]])
```
